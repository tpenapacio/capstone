```python
import pandas as pd
```

```python
def make_numeric(file_name):
  """
  Encodes all responses exlcuding "NO SMS" as 1, otherwise 0

  Input: data_file: file name of non-encoded data file
  """
  data = pd.read_csv(file_name)
  response = (data[[i for i in data.columns if 'Day' in i]] != "NO SMS").astype(int)
  meta_data = data[[i for i in data.columns if 'Day' not in i]]
  res = pd.concat([meta_data, response], axis = 1)
  return res
```

```python
file_name = "data/client_daily_reponse_details_clients.csv"
res = make_numeric(file_name)
res
```
